```python
import pandas as pd
df = pd.read_csv("revise_length_10.csv")
df
```

```python
temp_df = df.drop_duplicates('label')
vals = temp_df.label.tolist()
keys = temp_df.num_label.tolist()
convert_name = {}
for idx, v in enumerate(vals):
    convert_name[str(keys[idx])] = v

convert_name
```

```python
convert_dict = {}

for i in convert_name:
    convert_dict[i] = {
        'name':convert_name[i], 
        'sort':0
    }    
    
```

```python
import json
with open('./test_dict.json', 'w') as jf:
    json.dump(convert_dict, jf, ensure_ascii=False)
```
